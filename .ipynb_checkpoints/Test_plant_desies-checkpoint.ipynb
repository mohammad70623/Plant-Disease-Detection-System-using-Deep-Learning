Importing libraries

```python
import numpy as np 
import tensorflow as tf 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt 
import cv2
```

Loading Model

```python
model = tf.keras.models.load_model('../trained model/plant_desies_trained_model.keras')
```

```python
model.summary()
```

visualization single image of test sets

```python
image_path = "../Plant Disease data/test/PotatoEarlyBlight4.JPG"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Test image")
plt.xticks([])
plt.yticks([])
plt.show()
```

Testing model 

```python
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr]) #Converting single image to batch
print(input_arr.shape)
```

```python
predection = model.predict(input_arr)
predection, predection.shape
```

```python
result_index = np.argmax(predection)
print(result_index)
```
